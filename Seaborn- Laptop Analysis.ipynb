# HISTOGRAM

A histogram is a graph used to represent the frequency distribution of a few data points of one variable. Histograms often classify data into various bins or range groups and count how many data points belong to each of those bins

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['brand'], kde=True, bins=50, color='skyblue')
plt.title('Sales Distribution')
plt.xlabel('Brand')
plt.ylabel('Sales in crore')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df['display_size']), kde=True, bins=50, color='lightgreen')
plt.title('Sales Distribution (Log Transformed])')
plt.xlabel('Display_size')
plt.ylabel('Sales in crore')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['processor'], kde=True, bins=50, color='skyblue')
plt.title('processor value')
plt.xlabel('processor')
plt.ylabel('Sales in crore')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df['battery_capacity']), kde=True, bins=50, color='lightgreen')
plt.title('Battery Capacity')
plt.xlabel('Battery Capacity')
plt.ylabel('Sales in crore')

plt.tight_layout()
plt.show()
```

# PIE CHART

A pie chart (or a circle chart) is a circular statistical graphic which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents

```python
df['brand'].value_counts(normalize=True).head().plot(kind='pie',autopct='%0.2f%%')
```
